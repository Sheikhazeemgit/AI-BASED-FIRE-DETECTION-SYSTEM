```python
import threading
import smtplib
import tkinter as tk
import requests
import folium
from PIL import ImageTk, Image
from tkinter import messagebox
import ssl
import json
import pygame
import cv2

from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
from email.mime.base import MIMEBase
from email import encoders
import os
from tkinter import Tk, filedialog
from email_functions import send_mail_function
import shutil

# Create the Tkinter window
window = tk.Tk()
window.title("Fire Detection System")

# Set the window size and position
window_width = 800
window_height = 600

# Calculate the screen width and height
screen_width = window.winfo_screenwidth()
screen_height = window.winfo_screenheight()

# Calculate the window position to align it on the right side
window_x = screen_width - window_width
window_y = int((screen_height - window_height) / 2)

# Set the window geometry
window.geometry(f"{window_width}x{window_height}+{window_x}+{window_y}")

# Create a frame to hold the canvas and labels
frame = tk.Frame(window, bg="#f1f1f1")
frame.pack(pady=10, padx=10, fill=tk.BOTH, expand=True)

# Create a canvas to display video frames
canvas = tk.Canvas(frame, width=640, height=480, bg="white")
canvas.pack(side=tk.LEFT, padx=10, pady=10, fill=tk.BOTH, expand=True)


# Create a frame to hold the labels and entry fields
info_frame = tk.Frame(frame, bg="#f1f1f1")
info_frame.pack(side=tk.RIGHT, padx=10, fill=tk.Y)

# Create a label for displaying status and location
status_label = tk.Label(info_frame, text="Status: Waiting for fire detection", font=("Arial", 14), bg="#f1f1f1")
status_label.pack(pady=5, padx=5, anchor=tk.W)

# Create a label to display the confirmation message
confirmation_label = tk.Label(info_frame, text="", font=("Arial", 10), fg="green", bg="#f1f1f1")
confirmation_label.pack(pady=5, padx=5, anchor=tk.W)

# Create labels and entry fields for email credentials
sender_email_label = tk.Label(info_frame, text="Sender Email:", font=("Arial", 12), bg="#f1f1f1")
sender_email_label.pack(pady=5, padx=5, anchor=tk.W)

sender_email_entry = tk.Entry(info_frame, font=("Arial", 12), width=25)
sender_email_entry.pack(pady=5, padx=5, anchor=tk.W)

app_password_label = tk.Label(info_frame, text="Sender Password:", font=("Arial", 12), bg="#f1f1f1")
app_password_label.pack(pady=5, padx=5, anchor=tk.W)

app_password_entry = tk.Entry(info_frame, font=("Arial", 12), width=25, show="*")
app_password_entry.pack(pady=5, padx=5, anchor=tk.W)

recipient_email_label = tk.Label(info_frame, text="Recipient Email:", font=("Arial", 12), bg="#f1f1f1")
recipient_email_label.pack(pady=5, padx=5, anchor=tk.W)

recipient_email_entry = tk.Entry(info_frame, font=("Arial", 12), width=25)
recipient_email_entry.pack(pady=5, padx=5, anchor=tk.W)

message_label = tk.Label(info_frame, text="Message:", font=("Arial", 12), bg="#f1f1f1")
message_label.pack(pady=5, padx=5, anchor=tk.W)

message_text = tk.Text(info_frame, font=("Arial", 12), width=25, height=4)
message_text.pack(pady=5, padx=5, anchor=tk.W)

attachment_label = tk.Label(info_frame, text="Attachment:", font=("Arial", 12), bg="#f1f1f1")
attachment_label.pack(pady=5, padx=5, anchor=tk.W)

attachment_entry = tk.Entry(info_frame, font=("Arial", 10), width=20, state='readonly')
attachment_entry.pack(pady=5, padx=5, anchor=tk.W)

def browse_file():
    attachment_path = filedialog.askopenfilename(filetypes=[("PDF Files", "*.pdf")])
    if attachment_path:
        # Specify the directory to save the attachment
        save_directory = "attachments/"

        # Create the directory if it doesn't exist
        if not os.path.exists(save_directory):
            os.makedirs(save_directory)

        # Get the filename from the attachment path
        filename = os.path.basename(attachment_path)

        # Move the attachment to the save directory
        save_path = os.path.join(save_directory, filename)
        shutil.move(attachment_path, save_path)

        # Update the attachment entry field
        attachment_entry.config(state='normal')
        attachment_entry.delete(0, tk.END)
        attachment_entry.insert(0, save_path)
        attachment_entry.config(state='readonly')

# Create a button to browse for an attachment
browse_button = tk.Button(info_frame, text="Browse", command=browse_file, bg="#3498db", fg="white")
browse_button.pack(pady=5)
browse_button.config(relief=tk.FLAT, width=10)

# Load email credentials from the configuration file
# Load email credentials from the configuration file
config_file = "config.json"
try:
    with open(config_file, "r") as file:
        config_data = json.load(file)
        sender_email_entry.insert(0, config_data.get("sender_email", ""))
        recipient_email_entry.insert(0, config_data.get("recipient_email", ""))
        app_password_entry.insert(0, config_data.get("app_password", ""))
        message_text.insert(tk.END, config_data.get("message", ""))
        attachment_path = config_data.get("attachment_path", "")
        if attachment_path:
            attachment_entry.config(state='normal')
            attachment_entry.delete(0, tk.END)
            attachment_entry.insert(0, attachment_path)
            attachment_entry.config(state='readonly')
except FileNotFoundError:
    print("Configuration file not found.")

        

def save_email_credentials():
    # Retrieve the email credentials from the entry fields
    sender_email = sender_email_entry.get()
    app_password = app_password_entry.get()
    recipient_email = recipient_email_entry.get()
    message = message_text.get("1.0", tk.END)
    attachment_path = attachment_entry.get()

    # Check if any of the fields are empty
    if not sender_email or not app_password or not recipient_email or not message:
        messagebox.showerror("Error", "All email fields are mandatory.")
        return

    if not message.strip():
        messagebox.showerror("Error", "Message field is mandatory.")
        return

    if not attachment_path:
        messagebox.showerror("Error", "Attachment field is mandatory.")
        return 

    # Save the email credentials to the configuration file
    config_data = {
        "sender_email": sender_email,
        "recipient_email": recipient_email,
        "message": message,
        "attachment_path": attachment_path
    }
    with open(config_file, "w") as file:
        json.dump(config_data, file)
    print("Email credentials saved successfully.")
    # Update the confirmation label
    confirmation_label.config(text="Email credentials saved successfully")

    # Disable the email input fields
    sender_email_entry.config(state=tk.DISABLED)
    app_password_entry.config(state=tk.DISABLED)
    recipient_email_entry.config(state=tk.DISABLED)
    message_text.config(state=tk.DISABLED)
    attachment_entry.config(state=tk.DISABLED)
    browse_button.config(state=tk.DISABLED)
    save_button.config(state=tk.DISABLED)
    stop_button.config(state=tk.NORMAL)


# Create a button to save email credentials
save_button = tk.Button(info_frame, text="Save Email", command=save_email_credentials, bg="#3498db", fg="white")
save_button.pack(pady=10)
save_button.config(relief=tk.FLAT, width=20, height=2)




# Load the cascade classifier for fire detection
fire_cascade = cv2.CascadeClassifier('fire_detection_cascade_model.xml')

# Create a VideoCapture object
vid = cv2.VideoCapture(0)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

runOnce = False

pygame.init()
def play_alarm_sound():
    pygame.mixer.music.load('fire_alarm.mp3')
    pygame.mixer.music.play(-1)  # -1 plays the sound indefinitely
    print("Fire alarm initiated")

def send_mail_function(sender_email, app_password, recipient_email, message_text, attachment_path):
    try:
        # Create a multipart message
        message = MIMEMultipart()
        message["From"] = sender_email
        message["To"] = recipient_email
        message["Subject"] = "Fire Detection Alert"

        # Add message body
        message.attach(MIMEBase("text", "plain"))
        message.attach(MIMEText(message_text, "plain"))

        # Add attachment
        if attachment_path:
            filename = os.path.basename(attachment_path)
            with open(attachment_path, "rb") as attachment_file:
                part = MIMEBase("application", "octet-stream")
                part.set_payload(attachment_file.read())
            encoders.encode_base64(part)
            part.add_header("Content-Disposition", f"attachment; filename= {filename}")
            message.attach(part)

        # Connect to the SMTP server and send the email
        with smtplib.SMTP("smtp.gmail.com", 587) as server:
            server.starttls()
            server.login(sender_email, app_password)
            server.send_message(message)
        print("Alert email sent successfully to {}".format(recipient_email))
    except smtplib.SMTPException as e:
        print("Error sending email:", str(e))
        
        print(sender_email,app_password,recipient_email)

def fire_detection():
    global runOnce
    while True:
        Alarm_Status = False
        ret, frame = vid.read()

        if not ret:
            stop_fire_detection()  # Stop fire detection if camera is not accessible
            messagebox.showerror("Error", "Failed to capture video frame. Make sure the camera is connected and accessible.")
            print("Failed to capture video frame. Make sure the camera is connected and accessible.")
            break

        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        fire = fire_cascade.detectMultiScale(gray, 1.2, 5)

        for (x, y, w, h) in fire:
            cv2.rectangle(frame, (x-20, y-20), (x+w+20, y+h+20), (255, 0, 0), 2)

            if not Alarm_Status:
                status_label.config(text="Status: Fire detected")
                print("Fire alarm initiated")
                threading.Thread(target=play_alarm_sound).start()
                Alarm_Status = True
                send_mail_function(sender_email_entry.get(), app_password_entry.get(), recipient_email_entry.get(), message_text.get("1.0", tk.END), attachment_entry.get())
                
        # Display the resulting frame
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(img)
        img = ImageTk.PhotoImage(image=img)
        canvas.create_image(0, 0, anchor=tk.NW, image=img)

        # Check if the window is closed
        if cv2.waitKey(1) & 0xFF == ord('q'):
            stop_fire_detection()
            break

def start_fire_detection():
    # Check if the "Save Email" button is disabled
    if save_button['state'] == tk.NORMAL:
        messagebox.showerror("Error", "Please save email credentials first.")
        return
    
    global runOnce, vid
    runOnce = False
    
    # Create a VideoCapture object
    vid = cv2.VideoCapture(0)
    vid.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
    vid.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)
    
    # Update the status label
    status_label.config(text="Status: Fire detection started")
    
    # Disable the "Start Fire Detection" button
    start_button.config(state=tk.DISABLED)
    
    # Enable the "Stop Fire Detection" button
    stop_button.config(state=tk.NORMAL)
    
    # Start the fire detection in a separate thread
    threading.Thread(target=fire_detection).start()


def stop_fire_detection():
    global vid
    status_label.config(text="Status: Fire detection stopped")
    start_button.config(state=tk.NORMAL)
    stop_button.config(state=tk.DISABLED)
    vid.release()
    cv2.destroyAllWindows()
    pygame.mixer.music.stop()

# Create buttons to start and stop fire detection
start_button = tk.Button(window, text="Start Fire Detection", command=start_fire_detection, bg="#2ecc71", fg="white")
start_button.pack(pady=10)
start_button.config(relief=tk.FLAT, width=20, height=2)

stop_button = tk.Button(window, text="Stop Fire Detection", command=stop_fire_detection, bg="#e74c3c", fg="white")
stop_button.pack(pady=10)
stop_button.config(state=tk.DISABLED, relief=tk.FLAT, width=20, height=2)

# Start the Tkinter event loop
window.mainloop()
```
